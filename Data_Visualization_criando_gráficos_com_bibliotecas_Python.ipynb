```python
anos = list(map(str, range(1980, 2014))) #não considera o ultimo valor. entao ele pegara até 2013.
```

**range(1980, 2014):**

A função range() gera uma sequência de números.
O primeiro argumento (1980) é o início da sequência (inclusive).
O segundo argumento (2014) é o fim da sequência (exclusivo). Isso significa que a sequência irá até 2013.
Portanto, range(1980, 2014) gera uma sequência de números inteiros: 1980, 1981, 1982, ..., 2012, 2013.

**map(str, ...):**

Aplica a função str() (que converte para string) a cada número gerado por range().
Cada número inteiro (1980, 1981, etc.) será convertido em sua representação de string ('1980', '1981', etc.).

**list(...):**

Converte o objeto map resultante em uma lista.

**Linha Central Laranja (dentro da caixa): Mediana**

Esta linha representa a mediana dos dados de imigração. A mediana é o valor que divide o conjunto de dados ao meio, ou seja, 50% dos anos tiveram um número de imigrantes abaixo desse valor e 50% tiveram um número acima.

Neste gráfico, a mediana está em torno de 700-750 imigrantes. Isso significa que, na metade dos anos entre 1980 e 2013, o número de imigrantes do Brasil para o Canadá foi igual ou inferior a esse valor, e na outra metade, foi igual ou superior.

**A Caixa (Box): Intervalo Interquartil (IQR)**

A caixa em si (a parte retangular) representa o Intervalo Interquartil (IQR), que vai do primeiro quartil (Q1) ao terceiro quartil (Q3).

Borda Inferior da Caixa (Q1): É o 25º percentil. Significa que 25% dos anos tiveram um número de imigrantes abaixo ou igual a este valor. Neste gráfico, Q1 está aproximadamente em 400-450 imigrantes.

Borda Superior da Caixa (Q3): É o 75º percentil. Significa que 75% dos anos tiveram um número de imigrantes abaixo ou igual a este valor (ou 25% tiveram um número acima). Neste gráfico, Q3 está aproximadamente em 1000-1050 imigrantes.

A largura da caixa (Q3 - Q1) nos dá uma ideia da dispersão dos 50% centrais dos dados. Uma caixa mais estreita indica que os dados estão mais agrupados; uma caixa mais larga indica maior variabilidade.

**As "Hastess" (Whiskers): Intervalo Típico dos Dados**

As linhas verticais que se estendem para cima e para baixo da caixa são chamadas de "whiskers" (bigodes ou hastes). Elas geralmente se estendem até o menor e o maior valor dentro de 1.5 vezes o IQR a partir das bordas da caixa (Q1 e Q3).

A extremidade inferior da haste mostra o valor mínimo de imigração que não é considerado um valor atípico. Parece estar em torno de 150-200 imigrantes.

A extremidade superior da haste mostra o valor máximo de imigração que não é considerado um valor atípico. Parece estar por volta de 1700-1750 imigrantes.

Essas hastes indicam o intervalo onde a maioria dos dados "normais" ou "esperados" está concentrada.

**Os Pontos (Círculos): Outliers (Valores Atípicos)**

Os pontos individuais acima ou abaixo das hastes são os outliers (valores atípicos ou anômalos). São os anos em que o número de imigrantes foi significativamente maior (ou menor, se houvesse pontos abaixo da haste inferior) do que o padrão geral observado na maioria dos anos.

Neste gráfico, vemos dois outliers na parte superior, acima da haste superior, um por volta de 2150-2200 e outro em torno de 2500-2550 imigrantes. Isso sugere que houve anos específicos (entre 1980 e 2013) em que a imigração do Brasil para o Canadá foi excepcionalmente alta em comparação com o resto do período.

fig, axs = plt.subplots(1, 2): Cria uma figura com 1 linha e 2 colunas de subplots. axs será um array unidimensional com dois objetos de eixo: axs[0] e axs[1].

```python
# Podemos começar analisando os 4 maiores países da América do Sul: Brasil, Colômbia, Argentina e Peru.
# Nosso objetivo é comparar as tendências em uma única figura, utilizando subplots
```

```python
fig,axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace = 0.5, wspace=0.5) #dando espaçamento entre os graficos
fig.suptitle('Imigração 4 maiores países da América do Sul\n 1980 á 2013')

#0,0 corresponde na posição 1

axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')
axs[0,0].set_xlabel('Ano')
axs[0,0].set_ylabel('Imigrantes')
axs[0,0].grid()
# axs[0,0].xaxis.set_major_locator(plt.MultipleLocator(5))

#0,1 corresponde na posição 2
axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colômbia')
axs[0,1].set_xlabel('Ano')
axs[0,1].set_ylabel('Imigrantes')
axs[0,1].grid()
# axs[0,1].xaxis.set_major_locator(plt.MultipleLocator(5))

#1,0 corresponde na posição 3
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_xlabel('Ano')
axs[1,0].set_ylabel('Imigrantes')
axs[1,0].set_title('Argentina')
axs[1,0].grid()
# axs[1,0].xaxis.set_major_locator(plt.MultipleLocator(5))


#1,1 corresponde na posição 4
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_xlabel('Ano')
axs[1,1].set_ylabel('Imigrantes')
axs[1,1].set_title('Peru')
axs[1,1].grid()
# axs[1,1].xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat: #vai criar um loop que vai percorrer todas as celulas
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ymim= 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymim,ymax) #ylim de linha e xlim de linha

plt.show()
```

O loc em pandas é usado para selecionar dados de um DataFrame ou Series com base em rótulos (labels) de linhas e colunas. Ele é puramente baseado em rótulos e aceita.

O método .ravel() é um método de arrays NumPy que "achata" o array, transformando-o em uma única dimensão.
No exemplo acima, se axs é um array (2, 2), axs.ravel() o transformaria em um array unidimensional:

[ax1, ax2, ax3, ax4]

```python
cores = [] #isso nos permite
for pais in america_sul_sorted.index:
  if pais == 'Brazil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5)) #criando um subplotde barras
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores )  #ao colocar o H criamos um grafico de barras na horizontal
ax.set_title('América do Sul: Brasil foi o 4 pais com mais imigrantes\n para o canadá no periodo de 1980 à 2013 ', fontsize= 18, loc ='left')
ax.set_xlabel('Número de Imigrante', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']): #para colocar textos juntamente as barras
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha= 'left', va='center') #ha determina a posição do texto, fontsize o tamanho, va determina o alinhamento vertical do texto

ax.set_frame_on(False) #para tirar a borda do grafico
ax.get_xaxis().set_visible(False) #para tirar o eixo x
ax.tick_params(axis='both', which = 'both', length=0) #para tirar os ticks(marcadores) do eixo y. both= tirar de todos


plt.show()
```
