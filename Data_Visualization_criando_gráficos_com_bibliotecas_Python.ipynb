```python
anos = list(map(str, range(1980, 2014))) #n√£o considera o ultimo valor. entao ele pegara at√© 2013.
```

**range(1980, 2014):**

A fun√ß√£o range() gera uma sequ√™ncia de n√∫meros.
O primeiro argumento (1980) √© o in√≠cio da sequ√™ncia (inclusive).
O segundo argumento (2014) √© o fim da sequ√™ncia (exclusivo). Isso significa que a sequ√™ncia ir√° at√© 2013.
Portanto, range(1980, 2014) gera uma sequ√™ncia de n√∫meros inteiros: 1980, 1981, 1982, ..., 2012, 2013.

**map(str, ...):**

Aplica a fun√ß√£o str() (que converte para string) a cada n√∫mero gerado por range().
Cada n√∫mero inteiro (1980, 1981, etc.) ser√° convertido em sua representa√ß√£o de string ('1980', '1981', etc.).

**list(...):**

Converte o objeto map resultante em uma lista.

```python
brasil = df.loc['Brazil', anos] #para pegar os anos apenas do brasil, com todos os anos
```

```python
brasil
```

```python
#agora vamos transformar em um data frame
```

```python
brasil_dic= {'ano':brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} #nomenando as colunas listando-as
dados_brasil = spd.DataFrame(brasil_dic)
```

```python
#tolist() = usado para criar uma lista
```

```python
dados_brasil #data frame formado
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(x,y)
```

```python
#melhorando a visualiza√ß√£o do gr√°fico
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(x,y)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show() #c√≥digo para apenas mostrar o grafico . tudo que tiver abaixo do show n√£o ser√° executado.
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(x,y)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.title('Imigrantes do Brasil ao longo dos anos')
plt.show()
```
