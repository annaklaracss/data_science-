```python
anos = list(map(str, range(1980, 2014))) #não considera o ultimo valor. entao ele pegara até 2013.
```

**range(1980, 2014):**

A função range() gera uma sequência de números.
O primeiro argumento (1980) é o início da sequência (inclusive).
O segundo argumento (2014) é o fim da sequência (exclusivo). Isso significa que a sequência irá até 2013.
Portanto, range(1980, 2014) gera uma sequência de números inteiros: 1980, 1981, 1982, ..., 2012, 2013.

**map(str, ...):**

Aplica a função str() (que converte para string) a cada número gerado por range().
Cada número inteiro (1980, 1981, etc.) será convertido em sua representação de string ('1980', '1981', etc.).

**list(...):**

Converte o objeto map resultante em uma lista.

```python
anos
```

```python
brasil = df.loc['Brazil', anos] #para pegar os anos apenas do brasil, com todos os anos
```

```python
brasil
```

```python
#agora vamos transformar em um data frame
```

```python
brasil_dic= {'ano':brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} #nomenando as colunas listando-as
dados_brasil = spd.DataFrame(brasil_dic)
```

```python
#tolist() = usado para criar uma lista
```

```python
dados_brasil #data frame formado
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(x,y)
```

```python
#melhorando a visualização do gráfico
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(x,y)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show() #código para apenas mostrar o grafico . tudo que tiver abaixo do show não será executado.
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(x,y)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.title('Imigrantes do Brasil ao longo dos anos')
plt.show()
```
